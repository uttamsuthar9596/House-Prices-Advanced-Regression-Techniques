## Convert Categorical Variables Into Numerical

```python
from pandas.api.types import CategoricalDtype
```

```python
house['MSZoning'] = house['MSZoning'].astype(CategoricalDtype(categories=['RL', 'RM', 'C (all)', 'FV', 'RH'], ordered = True)).cat.codes
house['Street'] = house['Street'].astype(CategoricalDtype(categories=['Pave', 'Grvl'], ordered = True)).cat.codes
house['LotConfig'] = house['LotConfig'].astype(CategoricalDtype(categories=['Inside', 'FR2', 'Corner', 'CulDSac', 'FR3'], ordered = True)).cat.codes
house['Neighborhood'] = house['Neighborhood'].astype(CategoricalDtype(categories=['CollgCr', 'Veenker', 'Crawfor', 'NoRidge', 'Mitchel', 'Somerst','NWAmes', 'OldTown', 'BrkSide', 'Sawyer', 'NridgHt', 'NAmes','SawyerW', 'IDOTRR', 'MeadowV', 'Edwards', 'Timber', 'Gilbert','StoneBr', 'ClearCr', 'NPkVill', 'Blmngtn', 'BrDale', 'SWISU','Blueste'], ordered = True)).cat.codes
house['Condition1'] = house['Condition1'].astype(CategoricalDtype(categories=['Norm', 'Feedr', 'PosN', 'Artery', 'RRAe', 'RRNn', 'RRAn', 'PosA','RRNe'], ordered = True)).cat.codes
house['Condition2'] = house['Condition2'].astype(CategoricalDtype(categories=['Norm', 'Artery', 'RRNn', 'Feedr', 'PosN', 'PosA', 'RRAn', 'RRAe'], ordered = True)).cat.codes
house['BldgType'] = house['BldgType'].astype(CategoricalDtype(categories=['1Fam', '2fmCon', 'Duplex', 'TwnhsE', 'Twnhs'], ordered = True)).cat.codes
house['HouseStyle'] = house['HouseStyle'].astype(CategoricalDtype(categories=['2Story', '1Story', '1.5Fin', '1.5Unf', 'SFoyer', 'SLvl', '2.5Unf','2.5Fin'], ordered = True)).cat.codes
house['RoofStyle'] = house['RoofStyle'].astype(CategoricalDtype(categories=['Gable', 'Hip', 'Gambrel', 'Mansard', 'Flat', 'Shed'], ordered = True)).cat.codes
house['RoofMatl'] = house['RoofMatl'].astype(CategoricalDtype(categories=['CompShg', 'WdShngl', 'Metal', 'WdShake', 'Membran', 'Tar&Grv','Roll', 'ClyTile'], ordered = True)).cat.codes
house['Exterior1st'] = house['Exterior1st'].astype(CategoricalDtype(categories=['VinylSd', 'MetalSd', 'Wd Sdng', 'HdBoard', 'BrkFace', 'WdShing','CemntBd', 'Plywood', 'AsbShng', 'Stucco', 'BrkComm', 'AsphShn','Stone', 'ImStucc', 'CBlock'], ordered = True)).cat.codes
house['Exterior2nd'] = house['Exterior2nd'].astype(CategoricalDtype(categories=['VinylSd', 'MetalSd', 'Wd Shng', 'HdBoard', 'Plywood', 'Wd Sdng','CmentBd', 'BrkFace', 'Stucco', 'AsbShng', 'Brk Cmn', 'ImStucc','AsphShn', 'Other', 'Stone', 'CBlock'], ordered = True)).cat.codes
house['Foundation'] = house['Foundation'].astype(CategoricalDtype(categories=['PConc', 'CBlock', 'BrkTil', 'Wood', 'Slab', 'Stone'], ordered = True)).cat.codes
house['BsmtCond'] = house['BsmtCond'].astype(CategoricalDtype(categories=['NoBase','TA', 'Gd' , 'Fa', 'Po','Ex'], ordered = True)).cat.codes
house['BsmtFinType1'] = house['BsmtFinType1'].astype(CategoricalDtype(categories=['NoBase','GLQ', 'ALQ', 'Unf', 'Rec', 'BLQ' , 'LwQ'], ordered = True)).cat.codes
house['BsmtFinType2'] = house['BsmtFinType2'].astype(CategoricalDtype(categories=['NoBase','GLQ', 'ALQ', 'Unf', 'Rec', 'BLQ',  'LwQ'], ordered = True)).cat.codes
house['Heating'] = house['Heating'].astype(CategoricalDtype(categories=['GasA', 'GasW', 'Grav', 'Wall', 'OthW', 'Floor'], ordered = True)).cat.codes
house['Functional'] = house['Functional'].astype(CategoricalDtype(categories=['Typ', 'Min1', 'Maj1', 'Min2', 'Mod', 'Maj2', 'Sev'], ordered = True)).cat.codes
house['GarageType'] = house['GarageType'].astype(CategoricalDtype(categories=['No','Attchd', 'Detchd', 'BuiltIn', 'CarPort','Basment','2Types'], ordered = True)).cat.codes
house['GarageQual'] = house['GarageQual'].astype(CategoricalDtype(categories=['No','TA', 'Fa', 'Gd',  'Ex', 'Po'], ordered = True)).cat.codes
house['GarageCond'] = house['GarageCond'].astype(CategoricalDtype(categories=['No','TA', 'Fa', 'Gd',  'Ex', 'Po'], ordered = True)).cat.codes
house['SaleType'] = house['SaleType'].astype(CategoricalDtype(categories=['WD', 'New', 'COD', 'ConLD', 'ConLI', 'CWD', 'ConLw', 'Con', 'Oth'], ordered = True)).cat.codes
house['SaleCondition'] = house['SaleCondition'].astype(CategoricalDtype(categories=['Normal', 'Abnorml', 'Partial', 'AdjLand', 'Alloca', 'Family'], ordered = True)).cat.codes
```

```python
def BsmtQual(cols):
    BsmtQual = cols[0]
    
    if BsmtQual == 'Ex':
        Ex = range(100,125)
        return random.choice(Ex)
    
    if BsmtQual == 'Gd':
        Gd = range(90,99)
        return random.choice(Gd)
    
    if BsmtQual == 'TA':
        Ta = range(80,89)
        return random.choice(Ta)
    
    if BsmtQual == 'Fa':
        Fa = range(70,79)
        return random.choice(Fa)
    
    if BsmtQual == 'Po':
        Po = range(1,69)
        return random.choice(Po)
    
    if BsmtQual == 'NoBase':
        return 0
    
    else:
        return BsmtQual 
```

```python
house['BsmtQual'] = house[['BsmtQual']].apply(BsmtQual, axis=1)
```

```python
house['BsmtQual'].dtype
```

```python
house = pd.get_dummies(house)
```

```python
house = house.drop(['Electrical_Mix','Utilities_NoSeWa'], axis=1)
```

## Split Data Into Train And Test

```python
X = house.drop('SalePrice', axis=1)
Y = house['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
model_mse = mean_squared_error(y_test, predict)
model_rmse = np.sqrt(model_mse)
model_rmse
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model,X,Y, cv=5)
scores.mean()
```
